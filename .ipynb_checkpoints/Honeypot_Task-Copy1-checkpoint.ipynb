```python
def parse_salary(salary_str):
    if pd.isna(salary_str):
        return None
    
#     # Remove non-numeric characters except for commas, periods, hyphens, and 'K' (case-insensitive)
    salary_str = re.sub(r"[^\d´'., ]", 'splittt', salary_str)
    salary_str = salary_str.split('splittt')
    list_parsed_values = []
    for salary_item in salary_str:
        salary_item = salary_item.strip(" .,")
        if salary_item.count('.') <= 2 and salary_item.count(',') <= 1 and len(salary_item) < 10:
            salary_item = re.sub(r'[^\d.,]', '', salary_item)
            salary_item = salary_item.replace('.', '').replace(',', '.')
            if salary_item != '':
                numeric_value = int(float(salary_item))
                if len(str(numeric_value)) <= 3 and numeric_value != 0:
                    list_parsed_values.append(numeric_value * 1000)
                else:
                    list_parsed_values.append(numeric_value)
        else:
            if ' ' in salary_item:
                salary_item_list = salary_item.split(' ')
                salary_item_list = [re.sub(r'[^\d]', '', special_salary_item) for special_salary_item in salary_item_list]
                salary_item_list = list(filter(None, salary_item_list))
                salary_item_list = [int(special_salary_item) for special_salary_item in salary_item_list]
                list_parsed_values.extend(salary_item_list)
    
            elif salary_item.count('.') > 2:
                salary_item_list = salary_item.split('.', 2)
                salary_item_list = [salary_item_list[0] + '.' + salary_item_list[1], salary_item_list[2]]
                salary_item_list = [re.sub(r'[^\d]', '', special_salary_item) for special_salary_item in salary_item_list]
                salary_item_list = list(filter(None, salary_item_list))
                salary_item_list = [int(special_salary_item) for special_salary_item in salary_item_list]
                list_parsed_values.extend(salary_item_list)

            
    if list_parsed_values:
        mean_salaries = int(sum(list_parsed_values ) / len(list_parsed_values ))
    else:
        return None
    return mean_salaries
```

```python
import re

def parse_salary(salary_str):
    if pd.isna(salary_str):
        return None
    
    # Remove non-numeric characters except for commas, periods, hyphens, and 'K' (case-insensitive)
    salary_str = re.sub(r'[^\d.,\-kK+ ]', '', salary_str)
    
    # Handle values like '70+' (ending with '+')
    if salary_str.endswith('+'):
        try:
            return int(float(salary_str.rstrip('+')) * 1000)
        except ValueError:
            return None

    # Handle values like '60 - 70' (2 or 3 digits before and after the range)
    if re.match(r'\d{2,3} - \d{2,3}', salary_str):
        values = re.findall(r'\d{2,3}', salary_str)
        numeric_values = [float(val) for val in values if val.isdigit()]
        if numeric_values:
            return int((sum(numeric_values) / len(numeric_values)) * 1000)

    # Handle values like '70-80' (2 or 3 digits before and after the range without space)
    if re.match(r'\d{2,3}-\d{2,3}', salary_str):
        values = re.findall(r'\d{2,3}', salary_str)
        numeric_values = [float(val) for val in values if val.isdigit()]
        if numeric_values:
            return int((sum(numeric_values) / len(numeric_values)) * 1000)

    # Handle salary ranges with 'K' (case-insensitive)
    if '-' in salary_str and re.search(r'[kK]', salary_str):
        values = re.findall(r'[\d.,]+', salary_str)
        values = [val.replace(',', '').replace('.', '') for val in values if val]  # Remove empty values
        if len(values) >= 2:
            numeric_values = [float(val) for val in values if val.isdigit()]
            if numeric_values:
                return int((sum(numeric_values) / len(numeric_values)) * 1000)
    
    # Handle salary ranges without 'K' (case-insensitive)
    if '-' in salary_str:
        values = re.findall(r'[\d.,]+', salary_str)
        values = [val.replace(',', '').replace('.', '') for val in values if val]  # Remove empty values
        if len(values) >= 2:
            numeric_values = [float(val) for val in values if val.isdigit()]
            if numeric_values:
                return int(sum(numeric_values) / len(numeric_values))
#     elif salary_str.count('.') > 2:
#         print(salary_str)
#         parts = salary_str.split('.', 2)
#         if len(parts) == 2:
#             try:
#                 return int((float(parts[0]) + float(parts[1])) / 2)
#             except ValueError:
#                 return None
        
    
    # Handle values with 'K' (case-insensitive)
    if re.search(r'[kK]', salary_str):
        try:
            return int(float(salary_str.replace('k', '').replace('K', '')) * 1000)
        except ValueError:
            return None
    
    # Handle values with commas or periods as decimal separators
    if '.' in salary_str or ',' in salary_str:
        try:
            return int(float(salary_str.replace(',', '').replace('.', '')))
        except ValueError:
            return None
    
    # Handle values without any special characters
    if salary_str.isdigit():
        return int(salary_str)

    return None  # Handle other cases as invalid values
```
